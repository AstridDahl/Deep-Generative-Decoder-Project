# KNN for normalized counts

```python
# the PC coordinates and the tissue
freq_to_knn = pd.read_csv("freq_to_knn.tsv", sep='\t')
```

```python
freq_to_knn
```

```python
# remove rows where the tissue type only appears once in the df
freq_to_knn = freq_to_knn[freq_to_knn['primaryTumorLocation'].map(freq_to_knn['primaryTumorLocation'].value_counts()) > 1]
```

```python
freq_to_knn
```

```python
train_data, val_test_data = train_test_split(freq_to_knn, test_size=0.2, random_state=seed, stratify=freq_to_knn['primaryTumorLocation'])
```

```python
train_data
```

```python
# remove rows where the tissue type only appears once in the df
val_test_data = val_test_data[val_test_data['primaryTumorLocation'].map(val_test_data['primaryTumorLocation'].value_counts()) > 1]
```

```python
label_counts = val_test_data['primaryTumorLocation'].value_counts()
label_counts
```

```python
validation_data, test_data = train_test_split(val_test_data, test_size=0.5, random_state=seed, stratify=val_test_data['primaryTumorLocation'])
```

```python
validation_data
```

```python
test_data
```

```python
# both train and val set make up the final train set, so as to match the DGD approach
concat = pd.concat([train_data, validation_data], axis=0)
concat
```

```python
merged_df = concat.sort_index(axis=0)
merged_df
```

```python
X_train = merged_df.iloc[:, :-1]
y_train = merged_df['primaryTumorLocation'] 
```

```python
X_train
```

```python
X_test = test_data.iloc[:, :-1]
y_test = test_data['primaryTumorLocation'] 
```

```python
# knn
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
predictions = knn.predict(X_test)
```

```python
# the prediction accuracy
sum(predictions == y_test)/len(y_test)
```

```python
sklearn_adjusted_rand_score = adjusted_rand_score(predictions, y_test) # Calculate Adjusted Rand Score
sklearn_adjusted_rand_score
```
